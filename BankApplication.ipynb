```
def deposit():
  while True:
    account_no = int(input("Please enter account number:"))
    if account_no not in bank_db['Acc_No.'].values:
      print('acc_no is not valid')
      continue
    else:
      break
  while True:
    amount = int(input("Please enter valid amount (deposit): "))

    if amount >0:
      print('amount is valid')
      bank_db['balance'][(bank_db[bank_db['Acc_No.']==account_no].index.values[0])]+=amount
      print('balance has been updated', bank_db['balance'][(bank_db[bank_db['Acc_No.']==account_no].index.values[0])])
      break

    else:
      #raise Exception("Sorry, invalid amount, please enter valid amount")
      continue
```

```
def withdraw():
  while True:
    account_no = int(input("Please enter account number: "))

    if account_no not in bank_db['Acc_No.'].values:
      print('acc_no is not valid')
      continue
    else:
      break

  while True:
    amount = int(input("Please enter valid amount (withdraw): "))

    if amount >0:
      print('amount is valid')
      bank_db['balance'][(bank_db[bank_db['Acc_No.']==account_no].index.values[0])]-=amount
      print('balance has been updated', bank_db['balance'][(bank_db[bank_db['Acc_No.']==account_no].index.values[0])])
      break
      
    else:
      #raise Exception("Sorry, invalid amount, please enter valid amount")
      continue 
```

```
def balance_check():
  while True:
    account_no = int(input("Please enter account number: "))

    if account_no not in bank_db['Acc_No.'].values:
      print('acc_no is not valid')
      continue

    else:
      print('Current Balance is ', bank_db['balance'][(bank_db[bank_db['Acc_No.']==account_no].index.values[0])])
      break
```

```
if __name__ == "__main__":
  while True:
    option = int(input('Enter \n 1 for balance check \n 2 for deposit \n 3 for withdraw \n'))
    if option not in [1,2,3]:
      print('Invalid entry, please try again')
      continue
    else:
      if option == 1:
        balance_check()
      elif option == 2:
        deposit()
      elif option == 3:
        withdraw()
```
